```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2022_forbes_billionaires.csv")
df.head()
```

```python
df.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.sort_values(by= ["networth"], ascending=False).head(10)
plt.figure(figsize=(20,10))
sns.histplot(x = "name", hue= "networth", data = df)
plt.show()
```

```python
a = df["source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["skyblue", "yellowgreen", "blue", "red", "tomato"]
plt.figure(figsize=(5,5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0,0), 0.5, color = "white")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc("font", size = 12)
plt.title("Top 5 Domains to Become a Billionaire", fontsize = 20)
plt.show()
```

```python
a = df["industry"].value_counts().head()
index = a.index
industries = a.values
custom_colors = ["skyblue", "tomato", "red", "blue", "yellowgreen"]
plt.figure(figsize=(5,5))
plt.pie(industries, labels= index, colors=custom_colors)
central_circle = plt.Circle((0,0), 0.5, color = "white")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc("font", size =12)
plt.title("Top 5 Industries with Most Number of Billionaires", fontsize = 20)
plt.show()
```
